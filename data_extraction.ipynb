```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import alpaca_trade_api as tradeapi
from MCForecastTools import MCSimulation
```

```python
from dotenv import load_dotenv
load_dotenv()
```

```python
alpaca_api_key = os.getenv("ALPACA_API_KEY")
alpaca_secret_key = os.getenv("ALPACA_SECRET_KEY")
api = tradeapi.REST(
    alpaca_api_key,
    alpaca_secret_key,
    api_version="v2"
)
```

```python
ticker_sect = ["XLB", "XLP", "XLC", "XLV", "XLI", "XLK", "XLY", "XLU", "XLF", "XLE", "XLFE"]
timeframe = "1D"
start_date = pd.Timestamp("2000-01-01", tz="America/New_York").isoformat()
end_date = pd.Timestamp("2020-12-01", tz="America/New_York").isoformat()
```

```python
sector_data = api.get_barset(
    ticker_sect,
    timeframe,
    start=start_date,
    end=end_date
).df

sector_data.head()
```

```python
sector_data.to_csv("sector_data.csv")
```
